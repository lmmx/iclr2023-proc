```python
import pandas as pd
df = pd.read_csv("submissions.tsv", sep="\t", converters={"keywords": lambda x: x.strip("[]").replace("'","").split(",")})
```

```python
df.head()
```

```python
df.keywords[0][0]
```

```python
all_keywords = {}

def add_kw(keywords):
    for kw in keywords:
        kw = kw.strip().lower()
        if kw:
            all_keywords.setdefault(kw, 0)
            all_keywords[kw] += 1

df.keywords.apply(add_kw)
```

```python
all_keywords = dict(sorted(all_keywords.items(), key=lambda e: e[1], reverse=True))
```
